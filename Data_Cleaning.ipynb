```python
def ft_to_cm(height_str):
            # Split the string at the foot symbol (') and the inch symbol (")
            feet, inches = map(int, str(height_str).replace('"', '').split("'"))
            # Calculate total inches
            total_inches = (feet * 12) + inches
            # Convert inches to centimeters (1 inch = 2.54 cm)
            cm = round(total_inches * 2.54, 0)
            return int(cm)

data["Height"] = data["Height"].apply(ft_to_cm)
```

```python
data["Weight"]=data["Weight"].str.strip("kg")
```

```python
# Convert the 'Weight' column to a string type first to avoid errors
data['Weight'] = data['Weight'].astype(str)

# Now, apply your function to the string column
def lbs_to_kg(x):
    if "lbs" in x:
        lbs = x.replace("lbs", "")
        kilograms = round(int(lbs) / 2.205, 0)
        return int(kilograms)
    else:
        # If no "lbs" is found, assume it's already a number and convert
        return int(x)

data['Weight'] = data['Weight'].apply(lbs_to_kg)

# Display the result to verify the conversion
print(data['Weight'].head())
```

```python
def money_converter(x):
    # Convert input to string to handle all data types
    x = str(x)

    # Handle missing values
    if pd.isna(x) or x.strip() == '':
        return None

    # Remove the euro symbol
    x = x.replace("€", "")

    # Check for monetary scales in a logical order
    if "M" in x:
        x = x.replace("M", "")
        return int(float(x) * 1_000_000)
    elif "K" in x:
        x = x.replace("K", "")
        return int(float(x) * 1_000)
    else:
        # For values with no scale
        return int(float(x))

# Apply the function to each column separately and reassign correctly
# Assuming you have already defined the 'money_converter' function
data['Value'] = data['Value'].apply(money_converter)
data['Wage'] = data['Wage'].apply(money_converter)
data['Release Clause'] = data['Release Clause'].apply(money_converter)
```
